```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
euler = pd.read_csv('compTime_euler.csv', sep=',')
barneshut = pd.read_csv('compTime_barneshut.csv', sep=',')

fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(111)
ax1.set_xlabel("N")
ax1.set_ylabel("ms")
ax1.scatter(euler.where(euler.n < 10000).n, euler.where(euler.n < 10000).ms, c='b', marker="x", label='Euler')
ax1.scatter(barneshut.n,barneshut.ms, c='r', marker="x", label='Barnes-Hut')
ax1.legend()
fig.savefig('euler_barneshut_time_comparison1.pdf')
plt.show()
```
