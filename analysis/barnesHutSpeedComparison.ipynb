```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
euler = pd.read_csv('compTime_euler.csv', sep=',')
barneshut = pd.read_csv('compTime_barneshut.csv', sep=',')

euler_avg = euler.groupby('n').mean()
barneshut_avg = barneshut.groupby('n').mean()

euler_avg = np.column_stack((euler_avg.index.to_numpy(), euler_avg.to_numpy()))
barneshut_avg = np.column_stack((barneshut_avg.index.to_numpy(), barneshut_avg.to_numpy()))

euler_cropped = euler_avg[euler_avg[:,0] < 10000,:]

fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(111)
ax1.set_xlabel(r"$N$")
ax1.set_ylabel("ms")
ax1.set_title(r"Average computation time (1 iteration) for a varying number of bodies $N$")
ax1.scatter(euler_cropped[:,0], euler_cropped[:,1], c='b', marker="x", label='Euler')
ax1.scatter(barneshut_avg[:,0], barneshut_avg[:,1], c='r', marker="x", label=r'Barnes-Hut | $\theta=0.5$')
ax1.legend()
fig.savefig('euler_barneshut_time_comparison1.pdf')
plt.show()
```
