```python
import numpy as np
import pandas as pd

spherical = pd.read_csv('spherical.csv', sep=',')
spiral = pd.read_csv('spiral.csv', sep=',')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.scatter(spherical.x, spherical.y, 2, 'r', 'o')
plt.show()

plt.figure(figsize=(10,5))
plt.scatter(spherical.theta, spherical.r, 2, 'r', 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.scatter(spiral.x, spiral.y, 2, 'r', 'o')
plt.show()

plt.figure(figsize=(10,5))
plt.scatter(spiral.theta, spiral.r, 2, 'r', 'o')
plt.show()
```
