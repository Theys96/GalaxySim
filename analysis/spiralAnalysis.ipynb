```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotCartesian(name, points):
    plt.figure(figsize=(5,5))
    plt.title("%s\nCartesian coordinates" % name)
    plt.xlabel(r'$x$'); plt.ylabel(r'$y$');
    plt.scatter(points.x, points.y, 2, 'r', 'o')
    plt.show()

def plotPolar(name, points):
    plt.figure(figsize=(10,5))
    plt.title("%s\nPolar coordinates" % name)
    plt.xlabel(r'$\theta$'); plt.ylabel(r'$r$');
    plt.scatter(points.theta, points.r, 2, 'b', 'o')
    plt.show()
```

```python
circular = pd.read_csv('circular.csv', sep=',')
plotCartesian("Example circular galaxy", circular)
plotPolar("Example circular galaxy", circular)
```

```python
spiral = pd.read_csv('spiral.csv', sep=',')
plotCartesian("Example spiral galaxy", spiral)
plotPolar("Example spiral galaxy", spiral)
```
