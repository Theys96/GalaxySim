```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
barneshut = pd.read_csv('compTime_barneshut.csv', sep=',')
ms = pd.read_csv('compTime_ms.csv', sep=',')

barneshut_avg = barneshut.groupby('n').mean()
ms_avg = ms.groupby('n').mean()

barneshut_avg = np.column_stack((barneshut_avg.index.to_numpy(), barneshut_avg.to_numpy()))
ms_avg = np.column_stack((ms_avg.index.to_numpy(), ms_avg.to_numpy()))
```

```python
fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(111)
ax1.set_xlabel(r"$N$")
ax1.set_ylabel("ms")
ax1.set_title(r"Average computation time (1 iteration) for a varying number of bodies $N$"+"\nMost-Significant-Bodies Method")
ax1.scatter(ms_avg[:,0], ms_avg[:,1], c='r', marker="x", label='Most-Significant-Bodies')
fig.savefig('ms_time_comparison1.pdf')
plt.show()
```

```python
barneshut = pd.read_csv('compTime_barneshut.csv', sep=',')
ms = pd.read_csv('compTime_ms.csv', sep=',')

fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(111)
ax1.set_xlabel(r"$N$")
ax1.set_ylabel("ms")
ax1.set_title(r"Average computation time (1 iteration) for a varying number of bodies $N$")
ax1.scatter(barneshut_avg[:,0], barneshut_avg[:,1], c='b', marker="x", label=r'Barnes-Hut | $\theta=0.5$')
ax1.scatter(ms_avg[:,0], ms_avg[:,1], c='r', marker="x", label='Most-Significant-Bodies')
ax1.legend()
fig.savefig('ms_time_comparison2.pdf')
plt.show()
```
